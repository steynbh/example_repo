## Summation

In mathematics, we use $\sum$ to denote the sum of certain values

```python
# Our list of values
x = [5, 10, 3]

# Sum of all values
sum_total = 0
n = len(x) # we denote n as the number of elements in x
```

$\sum_{i = 0}^n x_i$

```python
for i in range(n):
    sum_total += x[i]

print(f"Our sum total is {sum_total}")
```

$\sum_{i = 0}^n x_i + 2$

```python
sum_total = 0 # Reset our sum_total back to 0
for i in range(n):
    sum_total += x[i] + 2

print(f"Our sum total is {sum_total}")
```
